```python
# Orignial code from https://www.deepseek.com, accessed on March 21, 2025
# When given a follow-up prompt of 'Generate python code for visualization of NetworkX graph object that color each nodes according to its attribute. Allow the user to choose a type of layout, have the name of each node shown on the drawn node, and show the legend.'

from matplotlib.lines import Line2D

def visualize_risk_graph(G, layout_type='spring'):
    """
    Visualize the Risk board graph with customizable layout and styling.

    Parameters:
        G (nx.Graph): The Risk board graph created using NetworkX.
        layout_type (str): The type of layout to use for the graph visualization.
                           Options: 'spring', 'circular', 'random', 'shell', 'spectral'.

    Returns:
        None: Displays the graph visualization.
    """
    # Define a color map for continents
    # A dictrionary that maps each continent to a specific color. This will be used to color the nodes based on their continent attribute
    continent_colors = {
        'North America': 'orange',
        'South America': 'yellow',
        'Europe': 'blue',
        'Africa': 'brown',
        'Asia': 'green',
        'Australia': 'purple'
    }

    # Assign colors to nodes based on their continent attribute
    # iterate over all nodes in the graph G. 
    # for each node, retieve the continenet attribute. Check the corresbonding continent color from the continent_colors dictrionary
    # The result is a list of colors, one for each node in the same order as the node.
    node_colors = [continent_colors[G.nodes[node]['continent']] for node in G.nodes]

    # Define layout options
    layout_options = {
        'spring': nx.spring_layout,
        'circular': nx.circular_layout,
        'random': nx.random_layout,
        'shell': nx.shell_layout,
        'spectral': nx.spectral_layout,
        'kamada_kawai' : nx.kamada_kawai_layout
    }

    # check if the user-specified layout_type is a valid key in layout_options dictionary
    # if it is in their, the corresponding layout function is called with the graph G as input, and the result is stored in pos (a dictrionary of node positions)
    if layout_type in layout_options:
        pos = layout_options[layout_type](G)
    # if the user input layout_type is not valid, a message will be printed, and the 'spring' layout will be used
    else:
        print(f"Layout type '{layout_type}' not found. Using 'spring' layout instead.")
        pos = layout_options['spring'](G)

    # Create the plot
    plt.figure(figsize=(12, 8))

    # Draw nodes with increased size and colored by continent
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500)

    # Draw edges
    nx.draw_networkx_edges(G, pos, alpha=0.5)

    # Draw labels (territory names) with increased font size on nodes
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    # Create custom legend handles
    # iterate over the items in continent_colors. for each item in continent_colors:
    # Creates a Line2D object, which is used to represent a line or marker in Matplotlib.
    # The [0], [0] part specifies the x and y coordinates of the line.
    # set label equal to the continent, the markerfacecolor equal to the assigned color. 
    legend_elements = [Line2D([0], [0], marker='o', color='w', label=continent,
                       markerfacecolor=color, markersize=10) for continent, color in continent_colors.items()]
    
    # Add the legend
    # use legend_elements for the legend
    plt.legend(handles=legend_elements, frameon=False, title='Continents', bbox_to_anchor=(1.05, 1))


    # Display the graph
    plt.title(f"Risk Board Graph ({layout_type.capitalize()} Layout)", fontsize=14)
    plt.axis('off')  # Turn off the axis
    plt.tight_layout()
    plt.show()

# Assuming G is the Risk board graph created earlier
visualize_risk_graph(G, layout_type='spring')
```
