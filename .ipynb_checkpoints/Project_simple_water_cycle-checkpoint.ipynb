### 1.2.1 Evaproation, Condensation, and Precipitation Model

Water cycle can be simplified into a simple circular cycle of evaproation, condensation, and precipitation. The rate of water traveling in these three stages are affected by multiple factors such as temperature, humidity, and wind speed. Since these factors change, there are no fix rate of water traveling around. 

I plan to create a compartmental model with three compartments, Evaporation, Condensation, and Precipitation. 

- Evaporation is when waters turns into vaports and evaporates into the atmosphere. 
- Condensation is when atmospher water evaporate and turn into clouds.
- Precipitation is when water from the could fall to earth.

water traveling through these three stages can be represented graphically as a compartmental model:


The ODE version of this, using the notation in the immediate model above and in the code below, is:

$\frac{dE}{dt} = -Ei*e + Pi*p,$

$\frac{dC}{dt} = -Ci*c + Ei*e,$

$\frac{dP}{dt} = -Pi*p + Ci*c.$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import solve_ivp
from IPython.display import display, clear_output, set_matplotlib_formats
import pandas as pd
```

The part is meant to show the cycle of water between the three stages for 100 day.

```python
# define the derivatives 
def derivative (t, y, e, c, p):
    Ei = y[0]
    Ci = y[1]
    Pi = y[2]
    
    dEdt = -Ei*e + Pi*p
    dCdt = -Ci*c + Ei*e
    dPdt = -Pi*p + Ci*c

    return(dEdt, dCdt, dPdt)
```

```python
# declare the rates of e (evaporation), c (condensation), p (precipitation)
e = 1/20 # rate of evaproation
c = 1/20 # rate of condensation
p = 1/20 # rate of precipitation

# declare parameters
# how much water in different stages initially.
E0 = 100 # liters
C0 = 0 # liters
P0 = 0 # liters

# total amount of water in the system
Total = E0+C0+P0
y0 = [E0, C0, P0]

#create time array
t  = np.arange(0,100,1)

# call solve_ivp
soln = solve_ivp(derivative, (0,500),y0, t_eval = t, args=(e, c, p))

# unpack solution
E  = soln.y[0, :]
C = soln.y[1, :]
P = soln.y[2, :]
```

```python
# plot evaproation, condensation, and precipitation graph
plt.figure(figsize=(8,8))
plt.subplot(311)
plt.plot(t, E, label='evaproation', color='#4730db')
plt.plot(t, C, label='condensation', color='#3080db')
plt.plot(t, P, label='precipitation',color='#2bccbc')
plt.xlabel('Day')
plt.ylabel('Amount of Water (L)')
plt.title('Water Cycle of Evaproation, Condensation and Precipitation Model')
plt.legend(loc=0)
```

This part is meant to see the water percent distribution in each carriers before and after 100 days.

```python
# finding the initial and final percentage of total water each carrier has.
e_percent = [(E0/Total)*100, (E[-1]/Total)*100]
c_percent = [(C0/Total)*100, (C[-1]/Total)*100]
p_percent = [(P0/Total)*100, (P[-1]/Total)*100]
```

```python
# Create a dictionary then convert it to a DataFrame 
df = pd.DataFrame(
           {'Category': ['initial', 'final'],
           'evaporation': e_percent,
           'condensation': c_percent,
           'precipitation': p_percent})

# Create the plot
fig, ax = plt.subplots()

# Plot the first segment Evaporation
bars_e = ax.bar(df['Category'], df['evaporation'], label='Evaporation', color='#4730db')

# Plot the second segment Condensation on top of Evaporation
bars_c = ax.bar(df['Category'], df['condensation'], bottom=df['evaporation'], label='Condensation', color='#3080db')

# Plot the third segment Precipitation on top of both Evaporation and Condensation
bars_p = ax.bar(df['Category'], df['precipitation'], bottom=df['evaporation'] + df['condensation'], label='Precipitation', color='#2bccbc')

# Add labels and title
ax.set_xlabel('Day')
ax.set_ylabel('Percentage (%)')
ax.set_title('Percent of Water in Different Compartments')
ax.legend(bbox_to_anchor=(1, 0.5))

# Orignial code from https://chatgpt.com/, accessed on November 30, 2024
# When given a follow-up prompt of 'how to shown the percentage on a stack bar graph' to ChatGPT 4o mini
# show the percentage
for bars in [bars_e, bars_c, bars_p]:
    for bar in bars:
        height = round(bar.get_height(), 2)
        # check that the percent is greater than 0.5%, else don't show it on the graph
        if height <= 0.5:
                continue
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_y() + height / 2

        ax.text(x, y, f'{height}%', ha='center', va='center', color='white', fontweight='bold')

# Show the plot
plt.show()
```

```python
def ecp_model_initial_value():
    # Define initial conditions for the problem
    initial_values = [[100, 0,0], [100, 100, 0], [10, 10, 10], [50, 75, 100]]
    
    # Set up the figure and axes
    fig, axs = plt.subplots(len(initial_values), 2, figsize=(10, 10))

    # loop through initial_values to generate final water distribution results for different initial values
    for i in range(len(initial_values)):
        # setting initial values to values in the ratesinitial_values list
        E0 = initial_values[i][0]
        C0 = initial_values[i][1]
        P0 = initial_values[i][2]
        # find total amount of water in the system
        Total = E0 + C0 + P0
        y0 = [E0, C0, P0]
        # generate time array
        t = np.arange(0, 100, 1)
    
        # Solve the differential equations
        soln = solve_ivp(derivative, (0, 500), y0, t_eval=t, args=(e, c, p))
        E = soln.y[0, :]
        C = soln.y[1, :]
        P = soln.y[2, :]
    
        # Calculate the percentage of water in each carrier at initial day and final day
        e_percent = [(E0 / Total) * 100, (E[-1] / Total) * 100]
        c_percent = [(C0 / Total) * 100, (C[-1] / Total) * 100]
        p_percent = [(P0 / Total) * 100, (P[-1] / Total) * 100]

        # Original code from https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html, accessed on November 30, 2024
        # Plot the evaporation, condensation, and precipitation curves, the first graph on the i row 
        axs[i, 0].plot(t, E, label='Evaporation', color='#4730db')
        axs[i, 0].plot(t, C, label='Condensation', color='#3080db')
        axs[i, 0].plot(t, P, label='Precipitation', color='#2bccbc')
        axs[i, 0].set_xlabel('Day')
        axs[i, 0].set_ylabel('Amount of Water')
        axs[i, 0].set_title(f'Water Cycle (Case {i + 1})')
        axs[i, 0].legend()
    
        # data frame of evaporation, condensation, and precipitation initial and final totoal water percent.
        df = pd.DataFrame({
            'Category': ['Initial', 'Final'],
            'evaporation': e_percent,
            'condensation': c_percent,
            'precipitation': p_percent
        })

        # plot stacked bar graph, the second graph on the i row 
        bars_e = axs[i, 1].bar(df['Category'], df['evaporation'], label=f'Evaporation (rate: {round(e,3)})', color='#4730db')
        bars_c = axs[i, 1].bar(df['Category'], df['condensation'], bottom=df['evaporation'], label=f'Condensation (rate: {round(c,3)})', color='#3080db')
        bars_p = axs[i, 1].bar(df['Category'], df['precipitation'], bottom=df['evaporation'] + df['condensation'], label=f'Precipitation (rate: {round(p,3)})', color='#2bccbc')

        # Add labels and title
        axs[i, 1].set_xlabel('Day')
        axs[i, 1].set_ylabel('Percentage (%)')
        axs[i, 1].set_title(f'Percent of Water in Different Compartments ({i + 1})')
        axs[i, 1].legend(bbox_to_anchor=(1, 1))
    
        # Add the percentage values on top of each segment
        for bars in [bars_e, bars_c, bars_p]:
            for bar in bars:
                height = round(bar.get_height(), 3)
                if height <=0.5:
                    continue
                x = bar.get_x() + bar.get_width() / 2
                y = bar.get_y() + height / 2
                axs[i, 1].text(x, y, f'{height}%', ha='center', va='center', color='white', fontweight='bold')

    plt.savefig('better_spaced_plot.png', dpi=300, bbox_inches='tight')
    #plt.savefig('tight_plot.png', bbox_inches='tight')
    plt.tight_layout()
    plt.show()
```

### 1.2.2 atmosphere, ocean, deepground water, and river model

A simple water cycle model contailing 4 compartments: atmosphere, ocean, deepground water, and river was created to resemble real life water cycle more. The rate of water staying at a compartment will be fixed, but by how much it goes to individual compartments would be randomlized. 

water traveling through four three stages can be represented graphically as a compartmental model:



The ODE version of this, using the notation in the immediate model above and in the code below, is:

$\frac{dO}{dt} = a\_i*a*a\_random + r\_i*r + g\_i*g*g\_random - o\_i*o,$

$\frac{dG}{dt} = a\_i*a*(1-a\_random) - g\_i*g*g\_random - g\_i*g*(1-g\_random),$

$\frac{dR}{dt} = g\_i*g*(1-g\_random) - r\_i*r,$

$\frac{dA}{dt} = o\_i*o - a\_i*a*a\_random - a\_i*a*(1-a\_random).$

The part is meant to show the cycle of water between the four carriers for several days.

```python
# define the rates
o = 1/(3000*12*30)       # rate of water flowing out of ocean
g = 1/(200*12*30)      # rate of water flowing out of groundwater
r= 1/(5*30)       # rate of water flowing out of river
a = 1/10         # rate of water flowing out of atmosphere

# Since atmosphere's water flows to two different carriers, I set factor of water flow of atmosphere that goes into one carrier
# the other carrier will receave (1-a_random) percent of water flow of atmosphere from total amound of water flow out of atmosphere
a_random = 0.4

# Since deep groundwater's water flows to two different carriers, I set factor of water flow of groundwater that goes into one carrier
# the other carrier will receave (1-a_random) percent of water flow of groundwater from total amound of water flow out of groundwater
g_random = 0.6

# define the derivatives 
def derivative_water(t, y, o, g, r, a, a_random, g_random):
    g_i = y[0]
    o_i = y[1]
    r_i = y[2]
    a_i = y[3]

    # Water flows between compartments
    atmosphere_to_ocean =a_i*a*a_random
    atmosphere_to_groundwater =a_i*a*(1-a_random)

    ocean_to_atmosphere = o_i*o

    river_to_ocean = r_i*r

    groundwater_to_ocean = g_i*g*g_random
    groundwater_to_river = g_i*g*(1-g_random)

    dOdt = river_to_ocean +groundwater_to_ocean +atmosphere_to_ocean -ocean_to_atmosphere
    dGdt = atmosphere_to_groundwater -groundwater_to_ocean-groundwater_to_river
    dRdt = groundwater_to_river-river_to_ocean
    dAdt =  ocean_to_atmosphere -atmosphere_to_ocean-atmosphere_to_groundwater

    return dOdt, dGdt, dRdt, dAdt
```

```python
# definal initial water amount in each carrier
groundwater_0 = 10
ocean_0 = 10
river_0 = 10
atmosphere_0 = 10

Total_amount = groundwater_0+ocean_0+river_0+atmosphere_0
y0 = [ocean_0, groundwater_0, river_0, atmosphere_0]

t  = np.arange(0,1000,1)

# call solve_ivp
soln = solve_ivp(derivative_water, (0,1000),y0, t_eval = t, args=(o, g, r, a, a_random, g_random))

# unpack solution
ocean_y  = soln.y[0, :]
groundwater_y = soln.y[1, :]
river_y = soln.y[2, :]
atmosphere_y = soln.y[3, :]
```

```python
plt.plot(ocean_y, label="ocean", color = '#4a32e3') #4a32e3
plt.plot(groundwater_y, label="groundwater", color = '#2f8ff5')
plt.plot(river_y, label="river", color = '#2fd7f5')
plt.plot(atmosphere_y, label="atmosphere", color = '#3ef7c0')

plt.title('Water Cycle of Ocean, Groundwater, River, and Atmosphere Model')
plt.xlabel('Day')
plt.ylabel('Amount of water (L)')
plt.legend(loc='center right', bbox_to_anchor=(0.95, 0.35))
plt.show()
```

This part is meant to show the water percent distribution in each carriers before and after several days.

```python
# finding the initial and final percentage of total water each carrier has
ocean_percent = [(ocean_0/Total_amount)*100, (ocean_y[-1]/Total_amount)*100]
groundwater_percent = [(groundwater_0/Total_amount)*100, (groundwater_y[-1]/Total_amount)*100]
river_percent = [(river_0/Total_amount)*100, (river_y[-1]/Total_amount)*100]
atmosphere_percent = [(atmosphere_0/Total_amount)*100, (atmosphere_y[-1]/Total_amount)*100]
```

```python
# Create dataframe
df = pd.DataFrame({
           'Category': ['initial', 'final'],
           'ocean': ocean_percent,
           'groundwater': groundwater_percent,
           'river': river_percent,
           'atmosphere': atmosphere_percent})

# Create the plot
fig, ax = plt.subplots()

# Plot the first segment ocean
bars_o = ax.bar(df['Category'], df['ocean'], label='ocean', color='#4a32e3')

# Plot the second segment groundwater on top of ocean
bars_g = ax.bar(df['Category'], df['groundwater'], bottom=df['ocean'], label='groundwater', color='#2f8ff5')

# Plot the third segment river on top of both ocean and groundwater
bars_r = ax.bar(df['Category'], df['river'], bottom=df['ocean'] + df['groundwater'], label='river', color='#2fd7f5')

# Plot the fourth segment atmosphere on top of both ocean, groundwater and river
bars_a = ax.bar(df['Category'], df['atmosphere'], bottom=df['ocean'] + df['groundwater'] + df['river'], label='atmosphere', color='#3ef7c0')

# Add labels and title
ax.set_xlabel('Day')
ax.set_ylabel('Percentage (%)')
ax.set_title('Percent of Water in Different compartments')
ax.legend(bbox_to_anchor=(1, 0.5))

# show the percentage
for bars in [bars_o, bars_g, bars_r, bars_a]:
    for bar in bars:
        height = round(bar.get_height(), 3)
        if height <= 0.5:
                continue
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_y() + height / 2

        ax.text(x, y, f'{height}%', ha='center', va='center', color='white', fontweight='bold')

# Show the plot
plt.show()
```

If the rate of two or more stages are the same, then they will have about the same percentage later (100 days later).

```python
ogra_model_initial_values()
```

The Atmosphere, Ocean, Deepground water and River Model is a compartmental model, modeling a simplified water cycle with four compartment. From the bar graph above we can see that the initial distribution of water affects the final distribution of water by a certain amount. Typically after 1000 days ocean will have the highest percentage of water and groundwater having the second highest percentage of water. River and Atmosphere have the smallest close to zero percentage. This result alines with the conclusion from the Evaporation, Condensation, and Precipitation Model. In the current model, ocean has the smallest flow rate, thus in most cases it has the biggest final water distribution. Groundwater has the second smallest flow rate, thus in most cases it has the second biggest final water distribution. Compare to ocean and groundwater flow rate, atmosphere and river have big flow rates, and the result is them having the smallest percentage of water by the end (1000 days).
For this model we can see that even if the rate of water flow of each carrier is the same, by changing the initial water distribution, the final (1000 days later) water distribution changes respectively. Thus for more complicated water cycle, initial water distribution and rate of water flow both matters.

### 1.4 Discussion and Conclusion

This model provide some insignt on the water cycle and the water distribution in different stages. But this model is simplified and may not truthfully reflex the real life water cycle. In the Evaporation, Condensation, and Precipitation Model, I have assumed through out the days the rate of water flow from each compartments would be the same. From this project I learned that if the water cycle is simple, circling one way, and fixed water flow rate, then no matter what kind of water distribution initially, the final distribution will be depended on the rate of water flow in each carrier. Smaller the rate of water flow compare to other carriers will make the carrier have a larger distribution of water by the end. Bigger the rate of water flow compare to other carriers will make the carrier have a smaller distribution of water by the end. If two carriers have the same water flow rate, then the two will have relatively same percentage of total water. These are supported by the graphs from Evaporation, Condensation, and Precipitation Model. For example, we say carrier A and carrier B are the only carriers in a simple circling water cycle with fixed water flow rates, and both have the same flow rate, if carrier A initially has 90% of all water and carrier B has 10%, after few days, carrier A and B will each have roughly 50 percentage of water. This pattern also shows in more complecated scenarios, but now initial water distribution among the carriers also matters.

The answer to my question is if the water cycle is simple, circling one way, and fixed water flow rate, then no matter what kind of water distribution initially, the final distribution will be depended on the rate of water flow in each carrier. This is because in my Evaporation, Condensation, and Precipitation Model, changing initial distribution does not affect final (after several days) water distribution among carriers. changing water flow rate of each carriers does affect the final water distribution among carriers.
For more complicated water cycle, both initial water distribution among carriers and carriers' water flow rate matters.
