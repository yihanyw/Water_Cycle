```python
def visualize_risk_graph(G, layout_type='spring'):
    # Define a colormap for continents
    colormap = cm.get_cmap('Set3')  # Use the 'tab10' colormap (10 distinct colors)
    
    # Get a list of unique continents
    continents = list(set(nx.get_node_attributes(G, 'continent').values()))
    
    # Assign a color to each continent using the colormap
    continent_colors = {continent: colormap(i) for i, continent in enumerate(continents)}

    # Assign colors to nodes based on their continent attribute
    node_colors = [continent_colors[G.nodes[node]['continent']] for node in G.nodes]

    # Define layout options
    layout_options = {
        'spring': nx.spring_layout,
        'circular': nx.circular_layout,
        'random': nx.random_layout,
        'shell': nx.shell_layout,
        'spectral': nx.spectral_layout,
        'kamada_kawai' : nx.kamada_kawai_layout
    }

    # Get the selected layout
    if layout_type in layout_options:
        pos = layout_options[layout_type](G)
    else:
        print(f"Layout type '{layout_type}' not found. Using 'spring' layout instead.")
        pos = layout_options['spring'](G)

    # Create the plot
    plt.figure(figsize=(12, 8))

    # Draw nodes with increased size and colored by continent
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1000)

    # Draw edges
    nx.draw_networkx_edges(G, pos, alpha=0.5)

    # Draw labels with increased font size
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    # Create a legend for the graph
    for continent, color in continent_colors.items():
        plt.scatter([], [], c=color, label=continent)  # Empty scatter plot for legend
    plt.legend(scatterpoints=1, frameon=False, title='Continents', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Display the graph
    plt.title(f"Risk Board Graph ({layout_type.capitalize()} Layout)", fontsize=14)
    plt.axis('off')  # Turn off the axis
    plt.tight_layout()
    plt.show()

# Example usage
# Assuming G is the Risk board graph created earlier
visualize_risk_graph(G, layout_type='kamada_kawai')
```
